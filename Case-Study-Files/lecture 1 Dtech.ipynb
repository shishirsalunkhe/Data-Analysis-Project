```python
def clean_size(x):
    if "k" in x :
        x = x.replace("k","") 
        return float(x) / 1024
    elif "M" in x :
        x = x.replace("M","") 
        return float(x)  
    elif x == "Varies with device":
        return np.nan
    else:
        return float(x)
```

```python
df["Size"]= df["Size"].map(clean_size)
```
