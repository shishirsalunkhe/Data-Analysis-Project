```python
job_counts = df['JobTitle'].value_counts().head(5)
job_counts.plot(kind='bar', color='skyblue')
```

```python
plt.scatter(df['BasePay'], df['OvertimePay'], alpha=0.5, color='blue')
plt.title('Scatter Plot of OvertimePay vs. BasePay')
```

```python
job_counts = df['JobTitle'].value_counts().head(10)
job_counts.plot(kind='bar', color='skyblue')
```

```python
average_basepay = df.groupby('Year')['BasePay'].mean()
average_basepay.plot(kind='line', marker='o', color='blue')
```

```python
average_totalpay = df.groupby('JobTitle')['TotalPay'].mean()
top_10_highest_paying = average_totalpay.nlargest(10)
top_10_highest_paying.sort_values().plot(kind='barh', color='skyblue')
```

```python
plt.boxplot(df['TotalPay'], vert=False)
```

```python
sns.pairplot(df[['BasePay', 'OvertimePay', 'OtherPay', 'TotalPay', 'TotalPayBenefits']])
plt.suptitle('Pair Plot of Employee Compensation Variables', y=1.02)
plt.show()
```

```python
average_overtimepay = df.groupby('Year')['OvertimePay'].mean()
average_overtimepay.plot(kind='line', marker='o', color='blue')
plt.title('Average OvertimePay by Year')
plt.xlabel('Year')
plt.ylabel('Average OvertimePay')
```

```python
police_officer_data = df[df['JobTitle'] == 'Police Officer']

plt.hist(police_officer_data['TotalPay'], bins=10, color='blue', alpha=0.7)
plt.title('Distribution of TotalPay for Police Officers')
plt.xlabel('TotalPay')
plt.ylabel('Frequency')
```

```python
plt.scatter(df['OvertimePay'], df['TotalPay'], alpha=0.5, color='blue')
plt.title('Scatter Plot of OvertimePay vs. TotalPay')
plt.xlabel('OvertimePay')
plt.ylabel('TotalPay')
plt.grid()
```
